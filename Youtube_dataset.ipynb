```python
import pandas as pd
```

```python
df = pd.read_csv("/content/youtubers_df.csv")
```

```python
###Data Exploration

#To show rows and columns

df.shape
```

```python
#To show rank of columns

df.columns.to_series().groupby(df.dtypes).groups
```

```python
#for username column has a blend of numbers and letters

import re
df['Username_blend'] = df['Username'].apply(lambda x: re.sub(r'(\d+)([a-zA-Z]+)', r'\1 \2', x))
display(df)
```

```python
#for a missing name which had to be filtered out

df = df[df['Username'] != 'Missing Name']
display(df)
```

```python
#the missing values

df.isna().sum()
```

```python
#to remove the missing value in categories column

df['Categories'].fillna(df['Categories'].mode()[0], inplace=True)
display(df)
```

```python
# the decimal point was removed to have a balanced whole number
import numpy as np

def correct_decimal(value):
    if isinstance(value, float) and not value.is_integer():
        return int(str(value).replace('.', ''))
    return int(value)

# Apply the correction function to the 'Visit' column
df['Visits'] = df['Visits'].apply(correct_decimal)

print("\nCorrected DataFrame:")
display(df)
```

```python
#to remove the link column

df = df.drop('Links', axis=1)
display(df)
```

```python
df.info()
```

```python
#correlation by heatmap

import matplotlib.pyplot as plt
import seaborn as sns

data = df[['Comments', 'Rank', 'Suscribers', 'Likes', 'Visits', 'Comments']]

# Create a heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.show()

##Comments and Likes: The moderate positive correlation between Comments and Likes implies that encouraging comments might also boost likes
##Likes and Comments (0.33) More comments are associated with more likes & Subscribers and Likes (0.21) More subscribers are associated with more likes.
```

```python
###Trend Analysis

#Bar chart for Categories and Subscribers
df_grouped = df.groupby('Categories')['Suscribers'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 7))
sns.barplot(x=df_grouped.index, y=df_grouped.values, alpha=0.8)
plt.ylabel('Total Subscribers', fontsize=12)
plt.xlabel('Categories', fontsize=12)
plt.title('Total Subscribers by Categories', fontsize=15)
plt.xticks(rotation='vertical')
plt.show()

##The significant skew towards "Música y baile" indicates a strong trend in content consumption towards music and dance-related videos. This could be due to the universal appeal of music and dance, transcending language and cultural barriers.
##Top categories are musica Y baile & peliculus,animacion & musica y baile,peliculus & noticias y politica
```

```python
# Bar chart for Likes and Categories
df_grouped_likes = df.groupby('Categories')['Likes'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 7))
sns.barplot(x=df_grouped_likes.index, y=df_grouped_likes.values, alpha=0.8)
plt.ylabel('Total Likes', fontsize=12)
plt.xlabel('Categories', fontsize=12)
plt.title('Total Likes by Categories', fontsize=15)
plt.xticks(rotation='vertical')
plt.show()

##The substantial lead of "Música y baile" (20 million likes) highlights a strong preference for music and dance content among audiences.
##Categories like "Videojuegos" and "Humor" (6 million likes) are also highly engaging, reflecting significant interest in gaming and comedy content.
##The categories "Vlogs diarios" (Daily Vlogs) and "Animación" (Animation) also show strong engagement(5 million likes)
```

```python
# Bar chart for Comments and Categories
df_grouped_comments = df.groupby('Categories')['Comments'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 7))
sns.barplot(x=df_grouped_comments.index, y=df_grouped_comments.values, alpha=0.8)
plt.ylabel('Total Comments', fontsize=12)
plt.xlabel('Categories', fontsize=12)
plt.title('Total Comments by Categories', fontsize=15)
plt.xticks(rotation='vertical')
plt.show()

##The "musica y baile" (Music and Dance) category has the highest number of comments, surpassing 500,000. This indicates a high level of engagement and interaction within this category.
#Followed by "videojuegos" (Video Games) with around 130,000 comments & "Humor" (Comedy) with approximately 120,000 comments & "Animación" (Animation) with close to 100,000 comments
##Categories with the lowest number of comments include "DIY y Life Hacks," "Mascotas" (Pets), and "Diseño/arte" (Design/Art), each with fewer than 10,000 comments.
```

```python
###Audiences Study


# Bar chart for Subscribers and Country
df_grouped_country = df.groupby('Country')['Suscribers'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 7))
sns.barplot(x=df_grouped_country.index, y=df_grouped_country.values, alpha=0.8)
plt.ylabel('Total Subscribers', fontsize=12)
plt.xlabel('Country', fontsize=12)
plt.title('Total Subscribers by Country', fontsize=15)
plt.xticks(rotation='vertical')
plt.show()

##The United States has the highest number of subscribers, with over 6 billion.
###India follows closely behind the United States with a total subscriber count also exceeding 6 billion so it is observed that there are regional preferences for specific content categories.
```

```python
# Bar chart for Likes and Country
df_grouped_likes_country = df.groupby('Country')['Likes'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 7))
sns.barplot(x=df_grouped_likes_country.index, y=df_grouped_likes_country.values, alpha=0.8)
plt.ylabel('Total Likes', fontsize=12)
plt.xlabel('Country', fontsize=12)
plt.title('Total Likes by Country', fontsize=15)
plt.xticks(rotation='vertical')
plt.show()

##the significant disparity in likes across different countries, with the United States leading by a wide margin, followed by India, and a substantial drop thereafter.
#The remaining countries have progressively fewer likes, indicating varying levels of engagement.
```

```python
# Bar chart for Comments and Country
df_grouped_comments_country = df.groupby('Country')['Comments'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 7))
sns.barplot(x=df_grouped_comments_country.index, y=df_grouped_comments_country.values, alpha=0.8)
plt.ylabel('Total Comments', fontsize=12)
plt.xlabel('Country', fontsize=12)
plt.title('Total Comments by Country', fontsize=15)
plt.xticks(rotation='vertical')
plt.show()

##The United States has the highest number of comments by a significant margin, exceeding 600,000 comments.
# India is the second-highest in terms of comments with approximately 250,000 comments.
#Brazil follows India with around 200,000 comments and a diverse set of countries from various regions around the world, showing a global distribution of comments, albeit with significant differences in counts.
```

```python
###Performances metrices

# Calculate the average of subscribers, likes, and comments
average_subscribers = df['Suscribers'].mean()
average_likes = df['Likes'].mean()
average_comments = df['Comments'].mean()

print(f"Average Subscribers: {average_subscribers}")
print(f"Average Likes: {average_likes}")
print(f"Average Comments: {average_comments}")
```

```python
###Content Categories

# Group the data by category and sum the subscribers
grouped_df = df.groupby('Categories')['Suscribers'].sum().reset_index()
grouped_df = grouped_df.sort_values(by='Suscribers', ascending=False)
print(grouped_df.to_string())

##
```

```python
###Content Categories
#plot distribution of content categories which categories have the highest number of streamers

import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(x='Categories', data=df)

plt.title('Distribution of Content Categories', fontsize=15)
plt.xlabel('Categories', fontsize=12)
plt.ylabel('Number of Streamers', fontsize=12)
plt.xticks(rotation='vertical')
plt.show()

# Get the number of streamers for each category
category_counts = df['Categories'].value_counts()
print('Categories with the highest number of streamers:')
for category, count in category_counts.items():
    if count == category_counts.max():
        print(f'- {category}: {count}')

##The most popular content categories are music and dance, video games, and animation. The least popular content categories are travel, science and technology, and health and fitness. The data suggests that viewers are most interested in content that is entertaining, engaging, and relatable.
```

```python
###Benchmarking

# Calculate the average of subscribers, likes, and comments
average_subscribers = df['Suscribers'].mean()
average_likes = df['Likes'].mean()
average_comments = df['Comments'].mean()

above_average_subscribers = df[df['Suscribers'] > average_subscribers]
above_average_likes = df[df['Likes'] > average_likes]
above_average_comments = df[df['Comments'] > average_comments]
above_average_visits = df[df['Visits'] > df['Visits'].mean()]

print(f"Number of streamers with above-average subscribers: {len(above_average_subscribers)}")
print(f"Number of streamers with above-average likes: {len(above_average_likes)}")
print(f"Number of streamers with above-average comments: {len(above_average_comments)}")
print(f"Number of streamers with above-average visits: {len(above_average_visits)}")
```
